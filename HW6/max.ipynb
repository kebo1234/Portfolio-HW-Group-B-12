```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data

df = pd.read_excel('momentum_data.xlsx', sheet_name='deciles (total returns)')

df

```

```python
D1 = -df.iloc[:, 1] + df.iloc[:, 10]
D3 = ( -df.iloc[:,1] - df.iloc[:,1] - df.iloc[:,2] + df.iloc[:,10] + df.iloc[:,8] + df.iloc[:,7] ) / 3
D5 = ( -df.iloc[:,4] - df.iloc[:,5] + df.iloc[:,6] - df.iloc[:,3] - df.iloc[:,2] - df.iloc[:,1] + df.iloc[:,10] + df.iloc[:,9] + df.iloc[:,8] + df.iloc[:,7] ) / 5


```

```python
factors_df = pd.read_excel('momentum_data.xlsx', sheet_name='factors (excess returns)')

mkt = factors_df['MKT']
smb = factors_df['SMB']
hml = factors_df['HML']
mkt
```

```python
combined_df = pd.DataFrame({
    'D1': D1,
    'D3': D3,
    'D5': D5,
    'mkt': mkt,
    'val': hml
})

combined_df
```

```python
D1mean = D1.mean() * 12
D3mean = D3.mean() * 12
D5mean = D5.mean() * 12

D1std = D1.std() * np.sqrt(12)
D3std = D3.std() * np.sqrt(12)
D5std = D5.std() * np.sqrt(12)

D1sharpe = D1mean / D1std
D3sharpe = D3mean / D3std
D5sharpe = D5mean / D5std

D1skewness = D1.skew()
D3skewness = D3.skew()
D5skewness = D5.skew()

D1kurtosis = D1.kurtosis()
D3kurtosis = D3.kurtosis()
D5kurtosis = D5.kurtosis()

D1mcorr = D1.corr(mkt)
D3mcorr = D3.corr(mkt)
D5mcorr = D5.corr(mkt)

D1hcorr = D1.corr(hml)
D3hcorr = D3.corr(hml)
D5hcorr = D5.corr(hml)

summary_df = pd.DataFrame({
    "Mean (Annualized)": [D1mean, D3mean, D5mean],
    "Vol": [D1std, D3std, D5std],
    "Sharpe Ratio": [D1sharpe, D3sharpe, D5sharpe],
    "Skewness": [D1skewness, D3skewness, D5skewness],
    "Kurtosis": [D1kurtosis, D3kurtosis, D5kurtosis],
    "Corr w/ MKT": [D1mcorr, D3mcorr, D5mcorr],
    "Corr w/ HML": [D1hcorr, D3hcorr, D5hcorr]
}, index=["D1", "D3", "D5"])

display(summary_df)
```

### (b)

Yes, using more extreme deciles does give a higher mean, which can be thought of as a more concentrated version of the factor, while at the same time having a higher vol, which I interpret to have a higher exposure to these equities' idiosyncracies.

AQR's retail product should probably be the D1 product because it has the highest sharpe & highest mean returns. Retail investors are not vol savvy & therefore likely are not aware of the D1/D5 tradeoff.
